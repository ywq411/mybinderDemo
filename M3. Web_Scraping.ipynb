## Now observe the URL:
- release_date — Shows only the movies released in a specific year.
- countries 
- count
- start: the number of movies since; this is where you can change the value of the parameter to further fetch movies that are not included in the list. 
- ref_ — Takes us to the the next or the previous page. The reference is the page we are currently on. adv_nxt and adv_prv are two possible values. They translate to advance to the next page, and advance to the previous page, respectively.

```python
# read in the text of an HTML 
# before 
url = "https://www.imdb.com/search/title?release_date=2018-01-01,2018-12-31&countries=us&count=250&start=1&ref_=adv_nxt"
response = get(url)
print(response.text[:200])
```

```python
# find all the movie containers. Each container has a movie. 
movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-advanced')
```

- Parse one movie, store all the information in a dictionary and save it to the disk. 

```python
movie = movie_containers[0]

movie_name = movie.h3.a.text

# year of production
movie_year = movie.h3.find('span', class_ = 'lister-item-year text-muted unbold')
movie_year = movie_year.text

# movie rating
movie_rating = movie.strong; 
movie_rating = float(movie_rating.text)

# movie metascore
movie_metascore = movie.find('span', class_="metascore favorable")
movie_metascore = int(movie_metascore.text)

# genres
movie_genres = movie.find("span", class_="genre")
movie_genres = movie_genres.text.split(',')
movie_genres = [genre.strip() for genre in movie_genres ]

# number of votes
movie_votes = movie.find("span", attrs={"name":"nv"})
movie_votes = int(movie_votes['data-value'])

# gross revenue of the movies
movie_gross = movie.find_all("span", attrs={"name":"nv"})
if len(movie_gross)>1:
    movie_gross = movie_gross[1]
    movie_gross = movie_gross['data-value']
    movie_gross = int(''.join(movie_gross.split(',')))

# who are the stars of the movies?
persons = movie.find_all('p', attrs = {"class":""})[0]
persons = persons.find_all('a')
persons = [p.text for p in persons]

one_movie = dict()
one_movie["name"] = movie_name
one_movie["year"] = movie_year
one_movie['rating'] = movie_rating
one_movie['metascore'] = movie_metascore
one_movie['genres'] = movie_genres
one_movie['votes'] = movie_votes
one_movie['gross'] = movie_gross
one_movie['stars'] = persons
one_movie
```

```python
movie_metascore = movie.find('span', class_="metascore mixed")
if movie_metascore != None:
    movie_metascore = int(movie_metascore.text)
else:
    movie_metascore = None
```
